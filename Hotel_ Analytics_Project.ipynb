```python
def get_dist_in_km(x):
    if x.split()[1] == 'm':
        return float(x.split()[0])/1000
    elif x.split()[1] == 'km':
        return x.split()[0]
    
df['Distance to Landmark'] = df['Distance to Landmark'].apply(get_dist_in_km)
```

```python
# Analysing the Hotel to get more customers.
# Taking example of  Hotel- Nature Zone Jungle Resort, what can be done to increase number of booking.
```

```python
def dist_check(x):
    if x <= 6:
        return "Very Near (0 Km - 6 Km]"
    elif x <= 12:
        return "Near (6 Km - 12 Km]"
    elif x <= 20:
        return "Far (12 Km - 20 Km]"
    else:
        return "Too Far"
    
def rating_check(x):
    if x >= 4.1:
        return "Better Ratings"
    else:
        return "Worser Ratings"
```

```python
df['Nearness_to_landmark']= df['Distance to Landmark(Km)'].apply(lambda x: dist_check(x))
df['Ratings_compared_to_nature_zone_hotel']= df['Rating'].apply(lambda x: rating_check(x))
```
